```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
sws = stopwords.words('english')
print (sws)
sws.remove('off')
sws.remove('on')
sws.remove('over')
sws.remove('only')
sws.remove('now')
sws.remove('up')
sws.remove('yourself')
sws.remove('before')
sws.remove('what')
sws.remove('some')
sws.remove('once')
sws.remove('just')
sws.remove('out')
sws.remove('all')
sws.remove('about')
sws.remove('most')
sws.remove('from')
sws.remove('other')
sws.remove('no')
sws.remove('with')
sws.remove('who')
sws.remove('how')
sws.remove('through')
sws.remove('should')
sws.remove('you')
sws.remove('our')
sws.remove('we')
sws.remove('your')
sws.remove('will')
sws.remove('can')
sws.remove('don\'t')
sws.remove('couldn\'t')
sws.remove('have')
sws.remove('has')
sws.remove('isn\'t')
sws.remove('both')
sws.remove('this')
sws.remove('these')
sws.remove('why')
sws.remove('that')
sws.remove('them')
sws.remove('here')
print (sws)
```

```python
punct = '.,:!?-'
dicts = [ {} for i in range(len(cts))]
#print (dicts)

for index, row in data_cut3.iterrows():

    if index > 0 and index % 100000 == 0:
        print (index)

    wordlist = row[1].split()
    wordlist2 = []
    for word in wordlist:
        word = word.lower() # make all words lowercase
        word = word.strip(punct) # remove punctuation
        word2 = '' 
        flag = False
        for c in word: # remove unicode characters, unless:
            if ord(c) <= 127 and flag == False:
                word2 = word2 + c
            if ord(c) > 127:
                flag = True
                if word2 == '': # the only character is unicode
                    word2 = c
                if (ord(word[len(word)-2]) == ord(c)) and (word[len(word)-1] == 't'): # if the word ends in 't, keep that
                    word2 = word2 + '\'t'
                continue
        word = word2
        if word == 'you\'ll': # remove you'll
            word = 'you'
        wordlist2.append(word)
        wordlist = set(wordlist2) # convert list to set: each word only appears once

    for word in sws: # remove stopwords
        if word in wordlist:
            wordlist.remove(word)
#    print (wordlist)

    for ind, ct in enumerate(cts): # build dictionary
        if ct in row[2]:
            for word in wordlist:
                if word in dicts[ind].keys():
                    dicts[ind][word] += 1
                else:
                    dicts[ind][word] = 1
```

```python
numbers = [13009567, 7284131, 1000444, 6999863, 1235362, 922195, 2854407, 1530893]
total = 34836862
f = open('/home/aleksey/Downloads/vocabulary_wordcounts_percentages.txt', 'w')
for ind, ct in enumerate(cts):
    print (ct, '-', len(dicts[ind].keys()))
    f.write(ct+' - '+str(len(dicts[ind].keys()))+'\n')
    dict_view = [ (value, key) for key, value in dicts[ind].items() ]
    dict_view.sort(reverse = True)
    for value, key in dict_view:
        print ("{:20s} {:20d} {:15.6f}".format(key, value, 100.*value/numbers[ind]))
        f.write("{:20s} {:20d} {:15.6f}\n".format(key, value, 100.*value/numbers[ind]))
    print()
    f.write('\n')
f.close()
```

```python
allwords = {}
for i in range(len(cts)):
#    print (i)
    for key, value in dicts[i].items():
        if key not in allwords.keys():
            allwords[key] = value
        else:
            allwords[key] += value
#        print (key, value, allwords[key])
```
